```python
energy = pd.merge(electricityperperson, oilperperson,  how='left', left_on=['country','year'], right_on = ['country','year'])
energy = pd.merge(energy, coalperperson,  how='left', left_on=['country','year'], right_on = ['country','year'])
energy.fillna(0, inplace=True)
energy['totalusageperperson'] = energy['electricityperperson'] + energy['oilperperson']+ energy['coalusageperperson']
energy.head()
```

```python
totalenergy = pd.merge(population, totalelectricity,  how='left', left_on=['Country Name','year'], right_on = ['country','year'])
totalenergy = pd.merge(totalenergy, totaloil,  how='inner', left_on=['country','year'], right_on = ['country','year'])
totalenergy = pd.merge(totalenergy, totalhydro,  how='inner', left_on=['country','year'], right_on = ['country','year'])
totalenergy.fillna(0, inplace=True)
totalenergy.head()
```

```python
totalenergy['TOTALPRODUCTION'] = totalenergy['totalelectricity'] + totalenergy['totaloilenergy']+ totalenergy['totalhydroenergy']
totalenergy.head()
```

```python
totalreserves = pd.merge(population, oilreserves,  how='left', left_on=['Country Name','year'], right_on = ['country','year'])
totalreserves = pd.merge(totalreserves, naturalgasreserves,  how='inner', left_on=['country','year'], right_on = ['country','year'])
totalreserves.fillna(0, inplace=True)
totalreserves['TOTALRESERVES'] = totalreserves['oilreserves'] + totalreserves['oilreserves']
totalreserves.head(10)
```

```python
TOTALREQUIREMENT = pd.merge(population, energy,  how='inner', left_on=['Country Name','year'], right_on = ['country','year'])
TOTALREQUIREMENT.fillna(0, inplace=True)
TOTALREQUIREMENT['TOTALREQT'] = TOTALREQUIREMENT['population'] * TOTALREQUIREMENT['totalusageperperson']
TOTALREQUIREMENT.head(10)
```

```python
ENERGYANALYSIS = pd.merge(TOTALREQUIREMENT, totalreserves,  how='inner', left_on=['Country Name','year'], right_on = ['country','year'])
ENERGYANALYSIS = pd.merge(ENERGYANALYSIS, totalenergy,  how='inner', left_on=['Country Name_x','year'], right_on = ['country','year'])
ENERGYANALYSIS.fillna(0, inplace=True)
ENERGY = ENERGYANALYSIS[['Country Name_x','year','population_x','TOTALREQT','TOTALRESERVES','TOTALPRODUCTION']]
ENERGY.head()
```
